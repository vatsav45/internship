```python
stopwords = stopwords.words('english')

stopwords=set(stopwords)
stopwords.update(["abbott","company",'usd','b'])
lemmatizer = WordNetLemmatizer()
```

```python
def datacleaning(text):
    k=''
    sent = re.sub('[^A-Za-z]',' ',text)
    sent = sent.lower()
    words = sent.split()
    words = [lemmatizer.lemmatize(word, pos='v') for word in words if word not in stopwords]
    text = ' '.join(words)
    return text
  
```

```python
GLOVE_DIR = "D:\Internship\\glove.6B"
EMBEDDING_DIM = 50
#doubt on embedding_dim
```

```python
def read_data(file_name):
  #Opening file
    with open(file_name,'r', encoding="utf8") as f:
      #Creating empty set and dictonary for vocab and word respectively
        word_vocab = set() 
        word2vector = {}
        #Iterating over each line of file
        for line in f:
            #Spliting lines
            line_ = line.strip() 
            #Splitting words
            words_Vec = line_.split()            
            word_vocab.add(words_Vec[0])
            word2vector[words_Vec[0]] = np.array(words_Vec[1:],dtype=float)
    print("Total Words in DataSet:",len(word_vocab))
    return word_vocab,word2vector
```

```python
vocab, w2v = read_data(os.path.join(GLOVE_DIR, "glove.6B.50d.txt"))
```

```python
# Vectorising text
# process the whole observation into single list
X_train_list=list(X_train)
X_test_list = list(X_test)
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train_list)
word_index = tokenizer.word_index
num_tokens = len(word_index) +1
```

```python
X_train_sequences = tokenizer.texts_to_sequences(X_train_list)
X_test_sequences  = tokenizer.texts_to_sequences(X_test_list)
```

```python
maxlen = 100

X_train = pad_sequences(X_train_sequences, padding='post', maxlen=maxlen)
X_test = pad_sequences(X_test_sequences, padding='post', maxlen=maxlen)
print(X_train[0, :])
```

```python
#Creating embeddding weight matrix
embedding_matrix = np.zeros((num_tokens, EMBEDDING_DIM))

for word, i in word_index.items():
    embedding_vector = w2v.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
```

```python
embedding_layer = tf.keras.layers.Embedding(num_tokens,
                                          EMBEDDING_DIM,
                                          weights=[embedding_matrix],
                                          input_length=maxlen,
                                          trainable=False)
```

```python
from tensorflow.keras.layers import Conv1D, Bidirectional, LSTM, Dense, Input, Dropout
from tensorflow.keras.layers import SpatialDropout1D
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
sequence_input = Input(shape=(maxlen, ), dtype='int32')
embedding_sequences = embedding_layer(sequence_input)
x = SpatialDropout1D(0.2)(embedding_sequences)
x = Conv1D(64, 5, activation='relu')(x)
x = Bidirectional(LSTM(64, dropout=0.2, recurrent_dropout=0.2))(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(512, activation='relu')(x)
outputs = Dense(1, activation='softmax')(x)
model = tf.keras.Model(sequence_input, outputs)
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model_fit = model.fit(X_train, y_train, batch_size=10, epochs=10,
                    validation_data=(X_test, y_test))
```
